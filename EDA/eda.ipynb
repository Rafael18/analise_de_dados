### Juntando 3 tabelas de uma unica vez com merge 

```python
df_churn = df_clientes.merge(df_servicos, on=["IDCliente"]).merge(df_contratos, left_on=["IDCliente"], right_on=["customerID"])
df_churn.info()
```

### Removendo a coluna customerID, pois já existe a coluna IDCliente

```python
df_churn.drop(["customerID"], inplace=True, axis=1)
```

### Tratar os valores (Métodos de exclusão)
Opções:

1. Remover a coluna onde tem valores ausente (Pior opção) - pois só são 11 registro, e o remover geral 
2. Remover a coluna se todos os valores forem ausentes
3. Remover linhas com valores ausentes (NaN)
4. Remover linhas com todos os valores ausentes (NaN)

**Nota:** Após métodos de exclusão existe outra seção por métodos de preenchimento 

```python
# Remover a coluna onde tem valores ausente (Pior opção) - pois só são 11 registro, e o remover 
df_churn.dropna(axis=1)
```

```python
# Remover a coluna se todos os valores forem ausentes
df_churn.dropna(axis=1, how="all")
```

```python
# Remover linhas com valores ausentes (NaN)
df_churn.dropna(axis=0)
```

```python
# Remover linhas com todos os valores ausentes (NaN)
df_churn.dropna(axis=0, how="all")
```

### Inputação de valores ausentes
Opções:
1. Preencher todos os valores ausentes por Zero
2. Preencher só determinada coluna com valores ausentes por 
3. Preencher os valores zerados pela média

```python
# 1. Preencher todos os valores ausentes por Zero
df_churn.fillna(0)
```

```python
# 2. Preencher só determinada coluna com valores ausentes por Zero
df_churn.fillna(value={"TotalCharges":0})
```

```python
# 3. Preencher os valores zerados pela média
df_churn.fillna(value={"TotalCharges": df_churn["TotalCharges"].mean()})
```
