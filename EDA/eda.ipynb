### Tratar os valores (MÃ©todos de exclusÃ£o)
OpÃ§Ãµes:

1. Remover a coluna onde tem valores ausente (Pior opÃ§Ã£o) - pois sÃ³ sÃ£o 11 registro, e o remover geral 
2. Remover a coluna se todos os valores forem ausentes
3. Remover linhas com valores ausentes (NaN)
4. Remover linhas com todos os valores ausentes (NaN)

**Nota:** ApÃ³s mÃ©todos de exclusÃ£o existe outra seÃ§Ã£o por mÃ©todos de preenchimento 

```python
# Remover a coluna onde tem valores ausente (Pior opÃ§Ã£o) - pois sÃ³ sÃ£o 11 registro, e o remover 
df_churn.dropna(axis=1)
```

```python
# Remover a coluna se todos os valores forem ausentes
df_churn.dropna(axis=1, how="all")
```

```python
# Remover linhas com valores ausentes (NaN)
df_churn.dropna(axis=0)
```

```python
# Remover linhas com todos os valores ausentes (NaN)
df_churn.dropna(axis=0, how="all")
```

### InputaÃ§Ã£o de valores ausentes
OpÃ§Ãµes:
1. Preencher todos os valores ausentes por Zero
2. Preencher sÃ³ determinada coluna com valores ausentes por 
3. Preencher os valores zerados pela mÃ©dia

```python
# 1. Preencher todos os valores ausentes por Zero
df_churn.fillna(0)
```

```python
# 2. Preencher sÃ³ determinada coluna com valores ausentes por Zero
df_churn.fillna(value={"TotalCharges":0})
```

```python
# 3. Preencher os valores zerados pela mÃ©dia
df_churn.fillna(value={"TotalCharges": df_churn["TotalCharges"].mean()})
```
