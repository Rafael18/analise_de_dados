### Primeira forma de juntar as tabelas
- df_servicos + df_clientes

```python
df_servicos.rename(columns={"customerID": "IDCliente"}, inplace=True)
df_servicos
```

### Realizando o merge das tabelas usando como chave o valor da coluna IDCliente

```python
df_temp = df_clientes.merge(df_servicos, on=["IDCliente"])
```

```python
df_temp.info()
```

```python
df_temp.head()
```

### Juntando as tebelas e atribuindo a uma tabela tamporaria
- Essas tabelas não possuem IDs identicos, então é preciso especificar os IDs de forma clara

```python
df_churn_temp = df_temp.merge(df_contratos, left_on=["IDCliente"], right_on=["customerID"])
```

```python
df_churn_temp.info()
```

### Juntando 3 tabelas de uma unica vez com merge 

```python
df_churn = df_clientes.merge(df_servicos, on=["IDCliente"]).merge(df_contratos, left_on=["IDCliente"], right_on=["customerID"])
df_churn.info()
```

### Removendo a coluna customerID, pois já existe a coluna IDCliente
- axis = 1 # está especificando que desejamos excluir a coluna

```python
df_churn.drop(["customerID"], inplace=True, axis=1)
```
